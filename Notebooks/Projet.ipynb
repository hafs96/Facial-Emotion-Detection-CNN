
### 4.3 Pr√©traitement des images <span id="pretraitement-img"></span>

Les √©tapes de pr√©traitement incluent :

- üîÑ **Conversion** de la cha√Æne de pixels en matrice 2D
- üßº **Normalisation** des valeurs de pixels entre 0 et 1
- üìè **Redimensionnement** (si n√©cessaire)
- üß™ **Encodage one-hot** pour les √©tiquettes (labels)
- üîÅ **Augmentation de donn√©es** (rotation, zoom, flip horizontal...) pour am√©liorer la robustesse du mod√®le

```python
image_size=(48,48)
# Convertir la colonne 'pixels' du DataFrame en une liste
# Chaque √©l√©ment de cette liste est une cha√Æne de pixels repr√©sentant une image
pixels = df['pixels'].tolist()

# D√©finir la taille cible des images (largeur et hauteur en pixels)
width, height = 48, 48

# Cr√©er une liste vide qui contiendra toutes les images trait√©es
faces = []

# Boucler sur chaque s√©quence de pixels (chaque ligne de la colonne 'pixels')
for pixel_sequence in pixels:
    # Diviser la cha√Æne de pixels par l'espace (' ') et convertir chaque valeur en entier
    face = [int(pixel) for pixel in pixel_sequence.split(' ')]

    # Convertir la liste de pixels en tableau numpy et redimensionner √† 48x48
    face = np.asarray(face).reshape(width, height)

    # Redimensionner l'image √† la taille souhait√©e (ici aussi 48x48), apr√®s conversion en uint8
    face = cv2.resize(face.astype('uint8'), image_size)

    # Ajouter l'image trait√©e √† la liste des visages, apr√®s conversion en float32
    faces.append(face.astype('float32'))

# Convertir la liste compl√®te de visages en un tableau numpy
faces = np.asarray(faces)

# Ajouter une dimension pour repr√©senter le canal de couleur (grayscale = 1 canal)
# Cela permet d'avoir la forme attendue par Keras/TensorFlow : (batch_size, height, width, channels)
faces = np.expand_dims(faces, -1)

# Encoder les √©tiquettes 'emotion' en format one-hot encoding
# Cela transforme les nombres (0 √† 6) en vecteurs binaires (0 ‚Üí [1,0,0,0,0,0,0])
emotions = pd.get_dummies(df['emotion']).to_numpy()
```

## üõ†Ô∏è 5. Approche technique <span id="approche"></span>

### 5.1 üß† Architecture du mod√®le (CNN) <span id="cnn"></span>

Pour la reconnaissance des √©motions faciales, nous avons choisi d‚Äôutiliser un **r√©seau de neurones convolutifs (CNN)**.  
Ce type de r√©seau est particuli√®rement efficace pour le traitement d‚Äôimages gr√¢ce √† sa capacit√© √† capturer des caract√©ristiques locales comme les **contours**, **textures** et **formes visuelles**.

üîß **Structure choisie :**  

Voici l'architecture CNN utilis√©e dans ce projet :


```python
# Importation des couches et mod√®les depuis tensorflow.keras 
from tensorflow.keras.layers import Input, Conv2D, Activation, Dropout
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import AveragePooling2D, BatchNormalization
from tensorflow.keras.layers import GlobalAveragePooling2D, Flatten
from tensorflow.keras.layers import MaxPooling2D, SeparableConv2D
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import CSVLogger, EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
import os
```

```python
# Configuration
num_epochs = 5
batch_size = 32
patience = 5
datasets = ['fer2013']
base_path = "./Models"

# D√©finition du mod√®le
input_shape = (48, 48, 1)
num_classes = 7

model = Sequential()
model.add(Convolution2D(filters=16, kernel_size=(7, 7), padding='same',
                        name='image_array', input_shape=input_shape))
model.add(BatchNormalization())
model.add(Convolution2D(filters=16, kernel_size=(7, 7), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(.5))

model.add(Convolution2D(filters=32, kernel_size=(5, 5), padding='same'))
model.add(BatchNormalization())
model.add(Convolution2D(filters=32, kernel_size=(5, 5), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(.5))

model.add(Convolution2D(filters=64, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Convolution2D(filters=64, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(.5))

model.add(Convolution2D(filters=128, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Convolution2D(filters=128, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(.5))

model.add(Convolution2D(filters=256, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Convolution2D(filters=num_classes, kernel_size=(3, 3), padding='same'))
model.add(GlobalAveragePooling2D())
model.add(Activation('softmax', name='predictions'))

# Compilation du mod√®le
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Boucle principale sur les datasets
for dataset_name in datasets:
    print('Entra√Ænement du dataset:', dataset_name)

    # Callbacks
    log_file_path = os.path.join(base_path, dataset_name + '_emotion_training.log')
    csv_logger = CSVLogger(log_file_path, append=False)
    early_stop = EarlyStopping('val_loss', patience=patience)
    reduce_lr = ReduceLROnPlateau('val_loss', factor=0.1, patience=int(patience / 4), verbose=1)
    trained_models_path = os.path.join(base_path, dataset_name + '_simple_cnn')
    os.makedirs(trained_models_path, exist_ok=True)
    model_names = os.path.join(trained_models_path, '{epoch:02d}-{val_loss:.2f}.keras')
    model_checkpoint = ModelCheckpoint(model_names, monitor='val_loss', verbose=1, save_best_only=True)
    my_callbacks = [model_checkpoint, csv_logger, early_stop, reduce_lr]

    
    train_faces = np.random.rand(1000, 48, 48, 1).astype('float32')  
    val_faces = np.random.rand(200, 48, 48, 1).astype('float32')

    train_emotions = np.random.randint(0, 7, size=(1000,))
    val_emotions = np.random.randint(0, 7, size=(200,))
    train_emotions = to_categorical(train_emotions, num_classes=7)
    val_emotions = to_categorical(val_emotions, num_classes=7)

    # Affichage des formes pour v√©rifier
    model.summary()
    print("train_faces.shape =", train_faces.shape)
    print("train_emotions.shape =", train_emotions.shape)

    # Entra√Ænement
    history = model.fit(
        x=train_faces,
        y=train_emotions,
        batch_size=batch_size,
        epochs=num_epochs,
        verbose=1,
        callbacks=my_callbacks,
        validation_data=(val_faces, val_emotions)
    )

    # Sauvegarde finale
    final_model_path = os.path.join(trained_models_path, 'final_model.h5')
    model.save(final_model_path)
    print(f"Mod√®le complet sauvegard√© sous : {final_model_path}")
```

- Convolution 7 : 128 filtres, noyau 3x3, activation ReLU

#### üß© Bloc 5

- Convolution 10 : 7 filtres, noyau 3x3  
- üìè GlobalAveragePooling : r√©duction spatiale vers un vecteur de taille 7

üîÑ **Sch√©ma g√©n√©ral de chaque bloc :**

- Une op√©ration de **pooling**  
- Un **dropout** pour limiter l‚Äôoverfitting

```python
def str_to_array(pixels_str):
    return np.array(pixels_str.split(), dtype=np.uint8).reshape(48, 48)

X = np.array(df['pixels'].apply(str_to_array).tolist())
y = pd.get_dummies(df['emotion']).values  # One-hot encoding des √©motions
```

```python
X = X.reshape(-1, 48, 48, 1)
X = X.astype('float32') / 255.0
```

```python
from sklearn.model_selection import train_test_split

# Diviser en train + test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Diviser train en train + validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, stratify=y_train, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
# Pr√©diction sur le jeu de test
y_pred = model.predict(X_test)  # <- ici, X_test est ton "test_faces"
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Afficher la pr√©cision globale
accuracy = accuracy_score(y_true_classes, y_pred_classes)
print(f"üéØ Accuracy sur le test : {accuracy * 100:.2f}%")

# Matrice de confusion
conf_matrix = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Col√®re', 'D√©go√ªt', 'Peur', 'Joie', 'Tristesse', 'Surprise', 'Neutre'],
            yticklabels=['Col√®re', 'D√©go√ªt', 'Peur', 'Joie', 'Tristesse', 'Surprise', 'Neutre'])
plt.xlabel('Pr√©diction')
plt.ylabel('V√©rit√©')
plt.title('Matrice de confusion')
plt.show()

# Rapport d√©taill√© par classe
class_report = classification_report(y_true_classes, y_pred_classes,
                                    target_names=['Col√®re', 'D√©go√ªt', 'Peur', 'Joie', 'Tristesse', 'Surprise', 'Neutre'])
print("üìã Rapport de classification :\n", class_report)
```
