## 🧠 2. État de l’art <span id="etat"></span>
### 🔍 2.1 Reconnaissance faciale : définition et applications <span id="def-app"></span>
La **reconnaissance faciale** est une technique d’analyse d’image permettant d’identifier ou de vérifier une personne à partir de ses traits faciaux, comme la distance entre les yeux, la forme du nez, la position des sourcils, etc.

**Applications principales :**
- 🔐 **Sécurité** : déverrouillage biométrique (ex. : smartphones), contrôle d’accès.
- 🎥 **Vidéosurveillance** : identification en temps réel.
- 🛍️ **Marketing** : analyse comportementale, publicités ciblées.
- 🏥 **Santé** : suivi de troubles neurologiques via les expressions.
- 🤖 **Assistants intelligents** : interactions empathiques avec l'utilisateur.

Grâce au *deep learning*, la reconnaissance faciale atteint aujourd’hui une grande précision, même en présence de variations (💡 luminosité, 🎥 angle, 😷 occlusion).

### 😊 2.2 Détection d’émotions faciales : état actuel <span id="emotion"></span>
La **détection des émotions faciales** vise à interpréter les expressions du visage humain pour déduire un état émotionnel, dans le domaine du *Affective Computing*.


### 4.3 Prétraitement des images <span id="pretraitement-img"></span>

Les étapes de prétraitement incluent :

- 🔄 **Conversion** de la chaîne de pixels en matrice 2D
- 🧼 **Normalisation** des valeurs de pixels entre 0 et 1
- 📏 **Redimensionnement** (si nécessaire)
- 🧪 **Encodage one-hot** pour les étiquettes (labels)
- 🔁 **Augmentation de données** (rotation, zoom, flip horizontal...) pour améliorer la robustesse du modèle

```python
image_size=(48,48)
# Convertir la colonne 'pixels' du DataFrame en une liste
# Chaque élément de cette liste est une chaîne de pixels représentant une image
pixels = df['pixels'].tolist()

# Définir la taille cible des images (largeur et hauteur en pixels)
width, height = 48, 48

# Créer une liste vide qui contiendra toutes les images traitées
faces = []

# Boucler sur chaque séquence de pixels (chaque ligne de la colonne 'pixels')
for pixel_sequence in pixels:
    # Diviser la chaîne de pixels par l'espace (' ') et convertir chaque valeur en entier
    face = [int(pixel) for pixel in pixel_sequence.split(' ')]

    # Convertir la liste de pixels en tableau numpy et redimensionner à 48x48
    face = np.asarray(face).reshape(width, height)

    # Redimensionner l'image à la taille souhaitée (ici aussi 48x48), après conversion en uint8
    face = cv2.resize(face.astype('uint8'), image_size)

    # Ajouter l'image traitée à la liste des visages, après conversion en float32
    faces.append(face.astype('float32'))

# Convertir la liste complète de visages en un tableau numpy
faces = np.asarray(faces)

# Ajouter une dimension pour représenter le canal de couleur (grayscale = 1 canal)
# Cela permet d'avoir la forme attendue par Keras/TensorFlow : (batch_size, height, width, channels)
faces = np.expand_dims(faces, -1)

# Encoder les étiquettes 'emotion' en format one-hot encoding
# Cela transforme les nombres (0 à 6) en vecteurs binaires (0 → [1,0,0,0,0,0,0])
emotions = pd.get_dummies(df['emotion']).to_numpy()
```

## 🛠️ 5. Approche technique <span id="approche"></span>

### 5.1 🧠 Architecture du modèle (CNN) <span id="cnn"></span>

Pour la reconnaissance des émotions faciales, nous avons choisi d’utiliser un **réseau de neurones convolutifs (CNN)**.  
Ce type de réseau est particulièrement efficace pour le traitement d’images grâce à sa capacité à capturer des caractéristiques locales comme les **contours**, **textures** et **formes visuelles**.

🔧 **Structure choisie :**  

Voici l'architecture CNN utilisée dans ce projet :


```python
# Importation des couches et modèles depuis tensorflow.keras 
from tensorflow.keras.layers import Input, Conv2D, Activation, Dropout
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import AveragePooling2D, BatchNormalization
from tensorflow.keras.layers import GlobalAveragePooling2D, Flatten
from tensorflow.keras.layers import MaxPooling2D, SeparableConv2D
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import CSVLogger, EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
import os
```

```python
# Configuration
num_epochs = 5
batch_size = 32
patience = 5
datasets = ['fer2013']
base_path = "./Models"

# Définition du modèle
input_shape = (48, 48, 1)
num_classes = 7

model = Sequential()
model.add(Convolution2D(filters=16, kernel_size=(7, 7), padding='same',
                        name='image_array', input_shape=input_shape))
model.add(BatchNormalization())
model.add(Convolution2D(filters=16, kernel_size=(7, 7), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(.5))

model.add(Convolution2D(filters=32, kernel_size=(5, 5), padding='same'))
model.add(BatchNormalization())
model.add(Convolution2D(filters=32, kernel_size=(5, 5), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(.5))

model.add(Convolution2D(filters=64, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Convolution2D(filters=64, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(.5))

model.add(Convolution2D(filters=128, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Convolution2D(filters=128, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(AveragePooling2D(pool_size=(2, 2), padding='same'))
model.add(Dropout(.5))

model.add(Convolution2D(filters=256, kernel_size=(3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Convolution2D(filters=num_classes, kernel_size=(3, 3), padding='same'))
model.add(GlobalAveragePooling2D())
model.add(Activation('softmax', name='predictions'))

# Compilation du modèle
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Boucle principale sur les datasets
for dataset_name in datasets:
    print('Entraînement du dataset:', dataset_name)

    # Callbacks
    log_file_path = os.path.join(base_path, dataset_name + '_emotion_training.log')
    csv_logger = CSVLogger(log_file_path, append=False)
    early_stop = EarlyStopping('val_loss', patience=patience)
    reduce_lr = ReduceLROnPlateau('val_loss', factor=0.1, patience=int(patience / 4), verbose=1)
    trained_models_path = os.path.join(base_path, dataset_name + '_simple_cnn')
    os.makedirs(trained_models_path, exist_ok=True)
    model_names = os.path.join(trained_models_path, '{epoch:02d}-{val_loss:.2f}.keras')
    model_checkpoint = ModelCheckpoint(model_names, monitor='val_loss', verbose=1, save_best_only=True)
    my_callbacks = [model_checkpoint, csv_logger, early_stop, reduce_lr]

    
    train_faces = np.random.rand(1000, 48, 48, 1).astype('float32')  
    val_faces = np.random.rand(200, 48, 48, 1).astype('float32')

    train_emotions = np.random.randint(0, 7, size=(1000,))
    val_emotions = np.random.randint(0, 7, size=(200,))
    train_emotions = to_categorical(train_emotions, num_classes=7)
    val_emotions = to_categorical(val_emotions, num_classes=7)

    # Affichage des formes pour vérifier
    model.summary()
    print("train_faces.shape =", train_faces.shape)
    print("train_emotions.shape =", train_emotions.shape)

    # Entraînement
    history = model.fit(
        x=train_faces,
        y=train_emotions,
        batch_size=batch_size,
        epochs=num_epochs,
        verbose=1,
        callbacks=my_callbacks,
        validation_data=(val_faces, val_emotions)
    )

    # Sauvegarde finale
    final_model_path = os.path.join(trained_models_path, 'final_model.h5')
    model.save(final_model_path)
    print(f"Modèle complet sauvegardé sous : {final_model_path}")
```

📥 **Entrée** : image de dimensions 48x48 pixels en niveaux de gris.

#### 🧩 Bloc 1  
- Convolution 1 : 16 filtres, noyau 3x3, activation ReLU  
- Convolution 2 : 16 filtres, noyau 3x3, activation ReLU  
- 📉 AveragePooling 1 : fenêtre 2x2  
- ❌ Dropout 1 : taux = 0.5

#### 🧩 Bloc 2  
- Convolution 3 : 32 filtres, noyau 3x3, activation ReLU  
- Convolution 4 : 32 filtres, noyau 3x3, activation ReLU  
- 📉 AveragePooling 2 : fenêtre 2x2  
- ❌ Dropout 2 : taux = 0.5

#### 🧩 Bloc 3  
- Convolution 5 : 64 filtres, noyau 3x3, activation ReLU  
- Convolution 6 : 64 filtres, noyau 3x3, activation ReLU  
- 📉 AveragePooling 3 : fenêtre 2x2  
- ❌ Dropout 3 : taux = 0.5

#### 🧩 Bloc 4  
- Convolution 7 : 128 filtres, noyau 3x3, activation ReLU  
- Convolution 8 : 128 filtres, noyau 3x3, activation ReLU  
- 📉 AveragePooling 4 : fenêtre 2x2  
- ❌ Dropout 4 : taux = 0.5

#### 🧩 Bloc 5  
- Convolution 9 : 256 filtres, noyau 3x3, activation ReLU  
- Convolution 10 : 7 filtres, noyau 3x3  
- 📏 GlobalAveragePooling : réduction spatiale vers un vecteur de taille 7  
- 🎯 Activation **Softmax** : prédiction des probabilités des émotions

🔄 **Schéma général de chaque bloc :**  
- Deux couches convolutionnelles successives  
- Une couche de **batch normalization**  
- Une fonction d’activation **ReLU**  
- Une opération de **pooling**  
- Un **dropout** pour limiter l’overfitting

```python
def str_to_array(pixels_str):
    return np.array(pixels_str.split(), dtype=np.uint8).reshape(48, 48)

X = np.array(df['pixels'].apply(str_to_array).tolist())
y = pd.get_dummies(df['emotion']).values  # One-hot encoding des émotions
```
